## Data visualization

### Useful functions for plots

```python
def plot_stacked_bars(dataframe, title_, size_=(18, 10), rot_=0, legend_="upper right"):
    """
    Plot stacked bars with annotations
    """
    ax = dataframe.plot(
        kind="bar",
        stacked=True,
        figsize=size_,
        rot=rot_,
        title=title_
    )

    # Annotate bars
    annotate_stacked_bars(ax, textsize=14)
    # Rename legend
    plt.legend(["Retention", "Churn"], loc=legend_)
    # Labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad=0.99, colour="white", textsize=13):
    """
    Add value annotations to the bars
    """

    # Iterate over the plotted rectanges/bars
    for p in ax.patches:
        
        # Calculate annotation
        value = str(round(p.get_height(),1))
        # If value is 0 do not annotate
        if value == '0.0':
            continue
        ax.annotate(
            value,
            ((p.get_x()+ p.get_width()/2)*pad-0.05, (p.get_y()+p.get_height()/2)*pad),
            color=colour,
            size=textsize
        )

def plot_distribution(dataframe, column, ax, ylim_=None, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"]==0][column],
    "Churn":dataframe[dataframe["churn"]==1][column]})
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')  
    if not(ylim_ == None):
        plt.ylim(0,ylim_)
    #print(dataframe[column])
    #print(temp)
```

### 1. Client dataset
### Churn rates of all companies

```python
churn = client_df[['id', 'churn']]
churn.columns = ['Companies', 'churn']
churn_total = churn.groupby(churn['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
#print(churn_percentage.transpose())
plot_stacked_bars(churn_percentage.transpose(), "Churn status", (5, 5), legend_="lower right")
```

#### Churn rates of companies that are or aren't also gas client 

```python
churn = client_df[['id', 'has_gas', 'churn']]

# Churn rates of companies which are also gas clients
gas = churn.loc[churn['has_gas'] == 't']
gas = gas[['id', 'churn']]
gas.columns = ['Companies have gas', 'churn']
churn_total = gas.groupby(gas['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
plot_stacked_bars(churn_percentage.transpose(), "Churn status", (5, 5), legend_="lower right")

# Churn rates of companies which are not gas clients
no_gas = churn.loc[churn['has_gas'] == 'f']
no_gas = no_gas[['id', 'churn']]
no_gas.columns = ['Companies do not have gas', 'churn']
churn_total = no_gas.groupby(no_gas['churn']).count()
churn_percentage = churn_total / churn_total.sum() * 100
#print(churn_percentage.transpose())
plot_stacked_bars(churn_percentage.transpose(), "Churn status", (5, 5), legend_="lower right")

# Number of copanies that are/aren't also a gas client
print("Number of gas clients: ", gas.count()[0])
print("Number of not gas clients: ", no_gas.count()[0])
```

- The churn rate of customers who are not a gas client is 23.2% higher than the ones who are also a gas client.
- There are 3.5 times more non-gas_client than gas_client.

```python
def plot_distribution_price(dataframe, column, ax, ylim_=None, bins_=50):
    """
    Plot variable distirbution in a stacked histogram of churned or retained company
    """
    # Plot the histogram
    dataframe[column].plot(kind='hist', bins=bins_, ax=ax, stacked=False)                          
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')  
    if not(ylim_ == None):
        plt.ylim(0,ylim_)
    #print(dataframe[column])
    #print(temp)
```

```python
fig, axs = plt.subplots(nrows=6, figsize=(18, 50))
i = 0

for column in ['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix']:
    plot_distribution_price(price_df, column, axs[i])
    i += 1
```

## Verifying the hypothesis

### Define the price sensitivity as the correlation of the price and the churn 

If the customer is most likely to churn when the price increases but stay when the price decreases or doesn't change, we say that the price sensitivity of this customer is high. Otherwise, if the customer churn when the price does not increase, then price is not the leading factor that affects the churn status, other hypothesis and strategies need to be investigated. 

### Merge the forcasted price data and the churn from the client dataset into the price dataset

```python
price_churn = pd.merge(price_df, client_df[['id', 'forecast_price_energy_off_peak', 'forecast_price_energy_peak', 'forecast_price_pow_off_peak', 'churn']], on='id')
price_churn.head()
```

### Calculate the correlation of the columns and examine the correlation of the churn and different types of prices

```python
corr = price_churn.corr()
#print(corr)

cmap = sns.diverging_palette(250, 10, as_cmap=True)

def magnify():
    return [dict(selector="th",
                 props=[("font-size", "7pt")]),
            dict(selector="td",
                 props=[('padding', "0em 0em")]),
            dict(selector="th:hover",
                 props=[("font-size", "12pt")]),
            dict(selector="tr:hover td:hover",
                 props=[('max-width', '200px'),
                        ('font-size', '12pt')])
]

corr.style.background_gradient(cmap, axis=1)\
    .set_properties(**{'max-width': '80px', 'font-size': '10pt'})\
    .set_caption("Hover to magify")\
    .set_precision(2)\
    .set_table_styles(magnify())
```

- The correlation is plotted in the cell above, red indicates significant correlation and the blue indicates insignificant correlation.
- As shown in the plot, the correlation between the churn and the price (the bottom row and the most right column) are not significant correlated, meaning the price sensitivity is small.
